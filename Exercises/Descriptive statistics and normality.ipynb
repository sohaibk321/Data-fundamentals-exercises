```python
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
normal1 = np.random.normal(5, 0.5, 100)
normal2 = np.random.normal(10, 1, 100)

plt.hist(normal1)
plt.axvline(normal1.mean(), color='r', linestyle='solid', linewidth=2)
plt.axvline(normal1.mean() + normal1.std(), color='r', linestyle='dashed', linewidth=2)
plt.axvline(normal1.mean() - normal1.std(), color='r', linestyle='dashed', linewidth=2)
plt.show()
```

```python
agg_normal = np.add(normal1, normal2)
plt.hist(agg_normal)
plt.axvline(agg_normal.mean(), color='r', linestyle='solid', linewidth=2)
plt.axvline(agg_normal.mean() + agg_normal.std(), color='r', linestyle='dashed', linewidth=2)
plt.axvline(agg_normal.mean() - agg_normal.std(), color='r', linestyle='dashed', linewidth=2)
plt.show()
```

```python
poisson = np.random.poisson(9,100)
plt.hist(poisson)
plt.axvline(poisson.mean(), color='r', linestyle='solid', linewidth=2)
plt.axvline(poisson.mean() + poisson.std(), color='r', linestyle='dashed', linewidth=2)
plt.axvline(poisson.mean() - poisson.std(), color='r', linestyle='dashed', linewidth=2)
plt.show()
```

```python
exponential = np.random.exponential(.5, size=100)
plt.hist(exponential)
plt.axvline(exponential.mean(), color='r', linestyle='solid', linewidth=2)
plt.axvline(exponential.mean() + exponential.std(), color='r', linestyle='dashed', linewidth=2)
plt.axvline(exponential.mean() - exponential.std(), color='r', linestyle='dashed', linewidth=2)
plt.show()
```

```python
gamma = np.random.gamma(7,3,100)
plt.hist(gamma)
plt.axvline(gamma.mean(), color='r', linestyle='solid', linewidth=2)
plt.axvline(gamma.mean() + gamma.std(), color='r', linestyle='dashed', linewidth=2)
plt.axvline(gamma.mean() - gamma.std(), color='r', linestyle='dashed', linewidth=2)
plt.show()

#gamma functions are continuous so maybe histograms are not the best way to visualize this. Maybe a line graph or
# a whisker-tail box
```

```python
binomial = np.random.binomial(50, .3, 100)
plt.hist(binomial)
plt.axvline(binomial.mean(), color='r', linestyle='solid', linewidth=2)
plt.axvline(binomial.mean() + binomial.std(), color='r', linestyle='dashed', linewidth=2)
plt.axvline(binomial.mean() - binomial.std(), color='r', linestyle='dashed', linewidth=2)
plt.show()
```

```python
rayleigh = np.random.rayleigh(9, 100)
plt.hist(rayleigh)
plt.axvline(rayleigh.mean(), color='r', linestyle='solid', linewidth=2)
plt.axvline(rayleigh.mean() + rayleigh.std(), color='r', linestyle='dashed', linewidth=2)
plt.axvline(rayleigh.mean() - rayleigh.std(), color='r', linestyle='dashed', linewidth=2)
plt.show()
```
